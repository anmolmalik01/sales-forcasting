```python
import numpy
import pandas as pd

# To plot
import matplotlib.pyplot as plt  
%matplotlib inline    
import matplotlib as mpl
import seaborn as sns

#For date-time
import math
from datetime import datetime
from datetime import timedelta

from sklearn.preprocessing import LabelEncoder
```

## Importing DataSets

```python
store_data = pd.read_csv('./dataset/stores.csv')
features_data = pd.read_csv('./dataset/features.csv')
train_data = pd.read_csv('./dataset/train.csv')
test_data = pd.read_csv('./dataset/test.csv')
```

```python
# function to get understanding of Dataset

def data_understanding(df):
    print(df.columns)
    print("==========================================")
    print(df.dtypes)
    print("==========================================")
    print(df.shape)
    print("==========================================")
    print(df.isnull().sum())
    print("==========================================")
    print(df.head())
    print("==========================================")
    print(df.describe())
```

### Data understanding

```python
data_understanding(store_data)
```

```python
data_understanding(features_data)
```

```python
data_understanding(train_data)
```

### Combining store, features and train to create a merged dataset

```python
df = train_data.merge(features_data, how="inner", on=['Store', 'Date']).merge(store_data, how="inner", on=["Store"])
```

```python
data_understanding(df)
```

# Cleaning merged df

### Our dataset has the following problems
1. There are 2 holiday columns
2. Weekly sales values is in negative
3. Markdown has lot of null values
4. Data is not normalized

```python
# dropping isholiday

df.drop('IsHoliday_y', inplace=True, axis="columns")
```

```python
df.shape
```

### Weekly sales

```python
(df['Weekly_Sales']<=0).sum()
```

```python
df.shape
```

```python
df = df.drop(df[df['Weekly_Sales'] <= 0].index)
```

```python
df.shape
```

### Markdowns

```python
df.isna().sum()
```

```python
# filling nan values with 0
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
print(df["Date"].min())
print(df["Date"].max())
```

```python
df.columns
```

```python
print(df["Date"].value_counts())
print(df["IsHoliday_x"].unique())
```

```python
holiday_dates = df[df['IsHoliday_x'] == True]


unique_holiday_dates = holiday_dates['Date'].unique()
unique_holiday_dates
```

## EDA

```python
x = df['Dept']
y = df['Weekly_Sales']
plt.figure(figsize=(15,5))
plt.title('Weekly Sales by Department')
plt.xlabel('Departments')
plt.ylabel('Weekly Sales')
plt.scatter(x,y)
plt.show()
```

```python
x = df['Store']
y = df['Weekly_Sales']
plt.figure(figsize=(15,5))
plt.title('Weekly Sales by Stores')
plt.xlabel('Stores')
plt.ylabel('Weekly Sales')
plt.scatter(x,y)
plt.show()
```

```python
df["Weekly_Sales"].dtypes
df["Date"].dtype
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month
```

```python
df.head()
```

```python
df_temp = df[['year', 'Weekly_Sales']].groupby(['year']).mean()
df_temp = df_temp.round(2)
df_temp
```

```python
df_temp = df.groupby('month')['Weekly_Sales'].mean()
df_temp
```

```python
x = ["2010", "2011", "2012"]
y = df_temp['Weekly_Sales']
plt.figure(figsize=(15,5))
plt.title('Sales per year')
plt.xlabel('years')
plt.ylabel('Weekly Sales')
plt.plot(x,y)
plt.show()
```

```python
# Grouping by 'year' and 'month' and calculating the mean of 'Weekly_Sales'
grouped_data = df.groupby(['month', 'year'])['Weekly_Sales'].mean().unstack()

# Plotting the grouped data
grouped_data.plot()

plt.xlabel('Month')
plt.ylabel('Mean Weekly Sales')
plt.title('Mean Weekly Sales by Month and Year')
plt.legend(title='Year')

plt.show()
```

```python
fuel_price = pd.pivot_table(df, values = "Weekly_Sales", index= "Fuel_Price")
fuel_price.plot()
```

```python
temp = pd.pivot_table(df, values = "Weekly_Sales", index= "Temperature")
temp.plot()
```

```python
CPI = pd.pivot_table(df, values = "Weekly_Sales", index= "CPI")
CPI.plot()
```

```python
unemployment = pd.pivot_table(df, values = "Weekly_Sales", index= "Unemployment")
unemployment.plot()
```
